```python
import numpy as np
```

```python
fname = "tester.npy"
creation = np.ones((10,10), dtype='float64')
np.save(fname, creation)
creation.shape, creation.dtype
```

```python
loaded = np.load(fname)
loaded.shape, loaded.dtype
```

```python
loadmem = np.lib.format.open_memmap(fname, mode='r', dtype=loaded.dtype, shape=loaded.shape)
loadmem.shape, loadmem.dtype
```

```python
loadmem
```

```python
loaded == loadmem
```
