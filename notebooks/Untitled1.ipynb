```python
def trial(self, stack=None, row=None, dimension=0, ylim=(None, None), xlim=(None, None)):
    """ Plots graph for a given dimension """
    
    if stack is None: stack = self.space.center[0]
    if row is None: row = self.space.center[1]
    if ylim is True: ylim = (0, None)
    if xlim is True: xlim = (0, None)
    
    # calcs
    x = self.space.x
    gx = self.sums[dimension][stack][row]
    ax = self.sums['a%s' % dimension][stack][row]
    aa = self.sums['abs'][stack][row]
    
    gxa = np.abs(gx)
    ea = aa-gxa
    ex = ax-gxa
    
    mass = self.sums['mass'][stack][row]
    
    references = {
        'gx': gx,
        'ex': ex,
        'ea': ea,
    }
    
    te = np.abs(np.log(ex/ea))**-0.5
    ta = np.log(ax/aa)**-1
    to = np.log(gx/aa)**-2
    tu = np.log(gx/ea)**-1
    
    tau = np.log(ax/gx)
    G = gxa*tau
    
    vel = (gx*x)**0.5
    
    attempts = {
        'velocity': vel,
        'tau': tau,
        'Gv': (G*x)**0.5,
        #'te': te,
        #'ta': ta,
        #'to': to,
        #'tau': tau,
        #'G': G,
        #'G2': G,
        #'snu': gxa*(np.log(ex/gxa))**2
    }
    
    
    fig, axes = plt.subplots(len(attempts)+1,1,figsize=(15,20))
    
    for i, key in enumerate(references.keys()):
        sns.lineplot(x=x, y=np.abs(references[key]), ax=axes[0], label=key).set(title='reference', xlim=xlim, ylim=ylim)
        
    #sns.lineplot(x=x, y=mass, ax=axes[1])
    
    for i, key in enumerate(attempts.keys()):
        sns.lineplot(x=x, y=np.abs(attempts[key]), ax=axes[i+1], color='orange', label=key).set(title=key, xlim=xlim, ylim=ylim)
        sns.lineplot(x=x, y=vel, ax=axes[i+1], color='purple', label='standard velocity')
    return attempts
```
