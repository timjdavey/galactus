```python
def field(self, stack, row, dimension, ylim=(None, None), xlim=(None, None)):
    """ Plots graph for a given dimension """
    fig, ax1 = plt.subplots(figsize=(15,5))
    ax2 = ax1.twinx()
    x = self.space.x
    
    g = sns.lineplot(x=x, y=self.sums[dimension][stack][row], ax=ax1, label='Dimensional gravity')
    sns.lineplot(x=x, y=self.sums['abs'][stack][row], ax=ax1, label='Absolute gravity')
    sns.lineplot(x=x, y=self.sums['mass'][stack][row], ax=ax2, label='Mass', color='grey')
    
    g.set(ylim=ylim, xlim=xlim)
```

```python
field(sim, 2, 1, 1)
```
