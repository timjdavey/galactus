# Double points

```python
points = (5,101) # height, width
centr = int((points[0]-1)/2)
offset = int(0.4*points[1])

left_point = np.zeros(points)
left_point[centr][0+offset] = 1
right_point = np.zeros(points)
right_point[centr][-1-offset] = 1

two_points = Space([left_point, right_point])
```

```python
two_points.quiver()
```

```python
two_points.line(ylim=(0,0.1))
```
